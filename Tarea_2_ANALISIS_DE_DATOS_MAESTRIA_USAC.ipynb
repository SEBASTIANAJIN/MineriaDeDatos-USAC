# IMPORTACIÓN DE LIBRERÍAS Y DATOS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount("/content/drive")
```

```python
dfp = pd.read_csv('/summer-products-with-rating-and-performance_2020-08.csv')

dfp
```

# EXPLORACIÓN DE DATOS



```python
dfp.info()

dfp.isnull().sum()
```

# LIMPIEZA DE DATOS Y CREACIÓN DE CATEGORÍAS


```python
## Establecimiento de género
dfp['title_orig'].unique()

dfp['title_orig'] = dfp['title_orig'].str.strip().str.lower()
dfp['tags'] = dfp['tags'].str.strip().str.lower()

dfp['genero 1'] = dfp['title_orig'].str.contains('lady|women|woman|girl|girly|female|wife|gentlewoman|dame|madam', case=False).replace({True: 'Women', False: ''})
dfp['genero 2'] = dfp['title_orig'].str.contains('men|man|male|gent|dude|guy|chap|fellow|sir|youth', case=False).replace({True: 'Men', False: ''})
dfp['genero 3'] = dfp['tags'].str.contains('lady|women|woman|girl|girly|female|wife|gentlewoman|dame|madam', case=False).replace({True: 'Women', False: ''})
dfp['genero 4'] = dfp['tags'].str.contains('men|man|male|gent|dude|guy|chap|fellow|sir|youth', case=False).replace({True: 'Men', False: ''})


dfp['Genero'] = np.where(dfp['genero 1'] == 'Women','Women',
                        np.where(dfp['genero 3'] == 'Women','Women',
                                 np.where(dfp['genero 2'] == 'Men','Men',
                                          np.where(dfp['genero 4'] == 'Men','Men',
                                                   np.where((dfp['genero 1'] == 'Women') & (dfp['genero 2'] == 'Men'),'Hola',
                                                            np.where((dfp['genero 3'] == 'Women') & (dfp['genero 4'] == 'Men'),'Hola','NA'))))))

dfp['Genero'].value_counts()
```

```python
## Establecimiento de tipo de artículo
dfp['title_orig'].unique()

dfp['title_orig'] = dfp['title_orig'].str.strip().str.lower()
dfp['tags'] = dfp['tags'].str.strip().str.lower()

dfp['b1'] = dfp['title_orig'].str.contains('pants|trouser|underpants|knicker|panties|bloomer|drawer|short|slack|bermuda|breeche|brief|chinos|cord|denim|jean|overall|digger|leggin|choke|snuff|gag|wheeze|suit|bikini', case=False).replace({True: 'Bottom', False: ''})
dfp['b2'] = dfp['tags'].str.contains('pants|trouser|underpants|knicker|panties|bloomer|drawer|short|slack|bermuda|breeche|brief|chinos|cord|denim|jean|overall|digger|leggin|choke|snuff|gag|wheeze|suit|bikini', case=False).replace({True: 'Bottom', False: ''})
dfp['t1'] = dfp['title_orig'].str.contains('top|shirt|dress|blouse|jersey|pullover|tunic|turtleneck|guimpe|middy|shirtwaist|garibaldi|halter|dickie|polo|jacket|coat|fur|parka|sheath|sweater|cardigan|jumper|camisole|vest', case=False).replace({True: 'Top', False: ''})
dfp['t2'] = dfp['tags'].str.contains('top|shirt|dress|blouse|jersey|pullover|tunic|turtleneck|guimpe|middy|shirtwaist|garibaldi|halter|dickie|polo|jacket|coat|fur|parka|sheath|sweater|cardigan|jumper|camisole|vest', case=False).replace({True: 'Top', False: ''})
dfp['s1'] = dfp['title_orig'].str.contains('shoe|boot', case=False).replace({True: 'Shoes', False: ''})
dfp['s2'] = dfp['tags'].str.contains('shoe|boot', case=False).replace({True: 'Shoes', False: ''})


dfp['TypeCloth'] = np.where(dfp['b1'] == 'Bottom','Bottom',
                        np.where(dfp['b2'] == 'Bottom','Bottom',
                                 np.where(dfp['t1'] == 'Top','Top',
                                          np.where(dfp['t2'] == 'Top','Top',
                                                   np.where(dfp['s1'] == 'Shoes','Shoes',
                                                            np.where(dfp['s2'] == 'Shoes','Shoes',
                                                                     np.where((dfp['b1'] == 'Bottom') & (dfp['t1'] == 'Top'),'Hola',
                                                                              np.where((dfp['b2'] == 'Bottom') & (dfp['t1'] == 'Top'),'Hola','Other'))))))))

dfp['TypeCloth'].value_counts()
```

```python
# Establecimiento de categoría

dfp['Categoria'] = dfp['Genero'] + ' - ' + dfp['TypeCloth']
dfp
```

# TRANSFORMACIÓN DE DATOS

### ANALISIS 1: TOP SELLERS POR CATEGORIA

```python
dftopseller = pd.DataFrame(dfp.groupby('Categoria')['units_sold'].sum())
dftopseller = dftopseller.sort_values(by = 'units_sold', ascending=False)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Categoria', y='units_sold', data=dftopseller, palette='Blues_d')
plt.title('Suma de Cantidades por Categoría')
plt.xlabel('Categoria')
plt.ylabel('units_sold')
plt.xticks(rotation=90)
plt.grid(axis='y')


## La categoria con más unidades vendidas es la sección de Top Women, seguido de bottom, por tanto, se vende en su mayoría ropa para dama
```

### ANALISIS 2: SENSIBILIDAD DE PRECIO

```python
dfp['FLAG_DESCUENTO'] = np.where(dfp['price'] < dfp['retail_price'],"Y","N")
dfp


dfdiscount = pd.DataFrame(dfp.groupby(['FLAG_DESCUENTO','Categoria'])['units_sold'].sum())
dfdiscount = dfdiscount.sort_values(by = 'units_sold', ascending=False)
dfdiscount
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Categoria', y='units_sold', hue = "FLAG_DESCUENTO", data=dfdiscount, palette='Blues_d')
plt.title('Suma de Cantidades por Categoría y Flag de descuento')
plt.xlabel('Flag descuento')
plt.ylabel('units_sold')
plt.xticks(rotation=90)
plt.grid(axis='y')

## Inicialmente, se identificó que no existe una mayor venta cuando el producto se encuentra en descuento.
## Adicionalmente, en la siguiente gráfica se visualiza que el prendas de dama se evidencia este patrón aunque en hombres, se identificó cierta sensibilidad.
```

### Análisis 3: Calidad Vs Ventas

```python
dfp['Rango_Rating'] = np.where(dfp['rating'] >= 5,'5',
                               np.where(dfp['rating'] >= 4,'4',
                                        np.where(dfp['rating'] >= 3,'3',
                                                 np.where(dfp['rating'] >= 2,'2',
                                                          np.where(dfp['rating'] >= 1,'1',0)))))

# Group by para cantidad de unidades vendidas
dfrating = pd.DataFrame(dfp.groupby('Rango_Rating')['units_sold'].sum())
dfrating = dfrating.sort_values(by = 'units_sold', ascending = False)
dfrating
# Group by para cantidad de productos
dfproducts = pd.DataFrame(dfp.groupby('Rango_Rating').size().reset_index(name='Contador'))
dfproducts = dfproducts.sort_values(by = 'Contador', ascending = False)
dfproducts

#merge
dffinal = pd.merge(dfrating,dfproducts, how = 'inner', on='Rango_Rating')
dffinal['Porcentaje'] = dffinal['units_sold']  / dffinal['Contador']
dffinal = dffinal.sort_values(by = 'Porcentaje', ascending = False)
dffinal
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Rango_Rating', y='Porcentaje', data=dffinal, palette='Blues_d')
plt.title('Cantidad de unidades vendidas por rating')
plt.xlabel('Rating')
plt.ylabel('Unidades vendidas x producto')
plt.xticks(rotation=90)
plt.grid(axis='y')
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Rango_Rating', y='units_sold', data=dfrating, palette='Blues_d')
plt.title('Suma de unidades vendidas por rating')
plt.xlabel('Rating')
plt.ylabel('units_sold')
plt.xticks(rotation=90)
plt.grid(axis='y')
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Rango_Rating', y='Contador', data=dfproducts, palette='Blues_d')
plt.title('Suma de productos por rating')
plt.xlabel('Rating')
plt.ylabel('#Productos')
plt.xticks(rotation=90)
plt.grid(axis='y')
```

```python
# A través de los gráficos mostrados con anterioridad, principalmente con el gráfico inicial,
# se puede observar que no existe una correlación entre el rating y la cantidad de unidades vendidas aún cuando se determina la cantidad vendida sobre el número de opciones con ese rating.
# Sin embargo, esto puede estar sesgado dado que en la mayoría de casos el rating 5 no es comunmente utilizado y se asocia a la perfección por lo que, omitiendo este fenómeno podríamos decir que si existe correlación.
# A mayor rating, mayor cantidad de unidades vendidas.
```
